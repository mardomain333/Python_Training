```python
#Filter()
def check_positive_no(num):
    return True if num>0 else False

lst=list(range(-10,10))
positivelist=list(filter(check_positive_no,lst))
print(positivelist)
```

```python
#Lambda -> Anonymous Function
negativelist=list(filter(lambda x:x<0,lst))
print(negativelist)
```

```python
#Map()
lst=list(range(1,11))
squared_list=list(map(lambda x:x**2,lst))
print(squared_list)
```

```python
#Reduce()
from functools import reduce          #in python 3 functools in library through which using reduce function

multiplied=reduce(lambda x,y:x*y,lst) #it work lyk accumulator

print(multiplied)
```

```python
#Keyword Argument

#kwargs all to pass keywords variable lenght of arguments to a function

def lst(**kwargs):
    if kwargs:
        print(f"Contact {kwargs['name']} {kwargs['mob']} {kwargs['add']} {kwargs['course']}")
lst(name='ram',mob=5892834598,add='Kerala',course='Btech')
```

```python
#Arbitrary Argument

def lst(*args):
    if args:
        for name in args:
            print(name)
lst('ram','das','rohit')
```
