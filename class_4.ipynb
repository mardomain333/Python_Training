## Generator and iterators in Python 
1.Iterator
   An iterator is an object that allows you to traverse through all the elements of a collection 
   (like lists or tuples) one at a time.
    keypoints:
    *implements __iter__() and __next__() methods .
    * raises StopIteration when no more data is available
    * used in loops like for loop

```python
#Example of iterator:
class MyNumbers:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num +=1
            return x
        else:
            raise StopIteration
myclass=MyNumbers()
myiter=iter(myclass)
for x in myiter:
    print(x)
```

```python
class Products:
    def __iter__(self):
        self.cart=['car','bike','shirt']
        self.i=0
        return self
    def __next__(self):
        if self.i<len(self.cart):
            x=self.cart[self.i]
            self.i+=1
            return x
        else:
            raise StopIteration
myclass=Products()
myiter=iter(myclass)
for x in myiter:
    print(x)
```

### Generator 
A Generator is a special ype of iterator that is defined using a function and 
'yeild ' keyword.
key points:
* uses yield instead of return
* Generates values lazily (one at a time)
* more memeory efficient than normal functions
  

```python
#Example of Generator:
import time
def my_generator():
     for i in range(1,6):
         yield i
start=time.time()
gen=my_generator()
for value in gen:
     print(value)
end=time.time()
print(end-start)
```

Iterator vs Generator 

Iterator :Requirs class with __iter__() and __next__()
Generator:Uses function with yield keyword.
Iterator :More code,manual implementation
Generator:Simple and concise implementation
Iterator :Slightly more control
Generator: More memory efficient



```python

def give_cart():
    cart=['car','bike','shirt']
    for c in cart:
        yield c
```

```python
class product:
    def __init__(self,cart=None):
        if cart==None:
            self.cart=[]
        else:
            self.cart=cart
    def __iter__(self):
        for c in self.cart:
            yield c
gokul=product(['shirt','pen','pencil'])
for c in gokul:
    print(c)
      
```
