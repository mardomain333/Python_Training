1. Matrix transpose using with and without built in function

```python
matrix=[
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]
transpose=[[row[i] for row in matrix ]for i in range(len(matrix)+1)]
print(transpose)
```

2: Assignment share elements without sharing address


```python
lst1=[1,2,3,4,5,5]
lst2=lst1.copy()
print("Lst1->",id(lst1))
print("lst2->",id(lst2))
```

3: How to reassign the data that is temporarily deleted

```python
lst={
    "name":'Ram',
    "age":22,
    "salary":45000
}
salary=lst.pop('salary')
print(lst)

#reassigning
lst['salary']=salary
print(lst)
```
