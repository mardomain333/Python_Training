### Assignment 1-> Matrix transpose using with and without built in function

```python
#without builtin function
matrix=[
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]
transpose=[[row[i] for row in matrix] for i in range(len(matrix)+1)]
print(transpose)

#with builtin function
import numpy as np
print(np.array(matrix).T)
```

### Assignment3 ->: How to reassign the data that is temporarily deleted

```python
lst={
    "name":'Ram',
    "age":22,
    "salary":45000
}
salary=lst.pop('salary')
print(lst)

#reassigning
lst['salary']=salary
print(lst)
```

### Assigment 2-> Where to use Map,Filter,Reduce

```python
# map() function is used to transform every elements,like it modifies the elements based
#on the given function

lst=[1,2,3,4,5]
lst1=list(map(lambda x:x**2,lst))
print(lst1)
```

```python
#filter() function is used to filter the elements from the list ,we only 
#keep the elements that matches with the condition 

lst2=list(filter(lambda x:x&1,lst)) # getting only the odd numbers
print(lst2)
```

```python
#Reduce() function help us to Collapse the list values into single value,used when we 
#need one final result
from functools import reduce
cumi_sum=reduce(lambda x,y:x+y,lst)
print(cumi_sum)
```
