```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Student_sheet_Output - Sheet1.csv')

# Question columns
question_cols = df.columns[1:]

# Total score
df["Total_Score"] = df[question_cols].sum(axis=1)

# Percentage
total_questions = len(question_cols)
df["Percentage"] = (df["Total_Score"] / total_questions) * 100


# Classification
def classify(student):
    if student >= 60:
        return "Advanced"

    else:
        return "Beginner"

df["Category"] = df["Percentage"].apply(classify)

result=df[["NAME", "Total_Score","Percentage", "Category"]]
result=result.sort_values(by="Category",ascending=False)
result.to_excel('GEN AI TEST2_RESULT.xlsx',index=False)
```

```python
category_count = df["Category"].value_counts()
# Plot graph
plt.figure()

plt.bar(category_count.index, category_count.values)

plt.title("Student Skill Level Distribution")
plt.xlabel("Category")
plt.ylabel("Number of Students")

plt.show()
```
