### 4.Copies & Views


```python
a=np.arange(10)
b=a[2:5]
c=a[4:]
print(id(a))
print(id(b))
print(id(c))
```

```python
np.may_share_memory(a,b)
```

```python
np.may_share_memory(b,c)
```

```python
np.may_share_memory(a,c)
```

```python
#[:::] this creates a view they shares the memory so the change will
#happens in the orginal one

b[1]=0
print(a)
```

```python
#change happended in a as well ,when modifying the view ,the original array is modifying as well
#c
```

```python
a=np.arange(10)
c=a[::2].copy() #force a copy & it will use another memory
#c=a[::2]
c
```

```python
np.may_share_memory(a,c)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

months=['jan','feb','mar','apr','may','jun']
traffic=[150,2,180,220,250,310]
plt.plot(months,traffic)
plt.xlabel('Month')
plt.ylabel('Monthly Traffic')
plt.title('Monthly Website Traffic')
plt.show()
```

```python
plt.plot(months,traffic,marker='^',markersize=15,color='green',linestyle='-.',linewidth=3)
plt.xlabel('Month')
plt.ylabel('Monthly Traffic(in Thousands)')
plt.title('Monthly Website Traffic')
plt.show()
```

```python
months=['jan','feb','mar','apr','may','jun']
product_a_revenue=[45,55,40,70,80,75]
product_b_revenue=[35,40,50,55,70,68]

plt.plot(months,product_a_revenue,marker='o',linestyle='-',color='blue',label='Product A')
plt.plot(months,product_b_revenue,marker='+',linestyle='--',color='violet',label='Product b')

plt.xlabel('months')
plt.ylabel('Monthly Revenue(in $1000)')
plt.title('Monthly Revenue Comparison')
plt.legend()
plt.show()
```

```python
overs=[5,10,15,20,25,30,35,40,45,50]
India_score=[45,102,148,180,210,240,265,280,298,309]
Pakistan_score=[30,75,140,190,230,240,260,280,290,293]
plt.plot(overs,India_score,marker='o',color='Blue',label='IND')
plt.plot(overs,Pakistan_score,marker='.',color='Green',label='PAK')
plt.xlabel('Overs')
plt.ylabel('Scores')
plt.title('ODI match')
plt.legend()
plt.savefig('image1')
plt.show()
```

### Types of Plot

1>Bar plot

```python
#sample data
catalog=['A','B','C','D','E']
values=[5,10,15,20,25]

#create a bar plot
plt.bar(catalog,values,color=['red','g','b','black'],width=0.8)
#add labels and a title
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar plot Example')

#Display the plot
plt.show()
```

```python
#sample data
catalog=['A','B','C','D','E']
values=[5,10,15,20,25]

#create a bar plot
plt.barh(catalog,values,color=['red','g','b','black'])
#add labels and a title
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar plot Example')

#Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('IPL.csv')
teams=df.team1.unique()
team=[team[:3] for team in teams]
#print(teams)
win=[]
#print(df.columns)
for name in teams:
    win.append(df['winner'].str.count(name).sum())
plt.bar(team,win,color='blue')
plt.xlabel('Teams')
plt.ylabel('No: of times Winners')
plt.title('Winners')
plt.show()
```
