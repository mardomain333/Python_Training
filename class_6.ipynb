```python
#Histogram
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=np.random.random_sample(5)
print(data)
plt.hist(data,color='g',edgecolor='yellow',bins=5)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram plot Example')
plt.show()
```

```python
#scatterplot

data=np.random.randn(10)

#sample data
x=[1,2,3,4,5]
y=[10,15,13,17,20]

plt.scatter(x,y,marker='o',color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot Example')
plt.show()
```

```python


categories=['Housing','Transportation','Food','Entertainment','Utilities']
expense=[1200,300,400,200,150]

plt.figure(figsize=(5,16))
plt.pie(expense,labels=categories,autopct='%2.2f%%')

plt.title('Montly Expense by Category')
plt.legend()

plt.show()
```

```python
#stack plot
```

```python
quarters=['Q1','Q2','Q3','Q4']
electronics=[10000,12000,11000,10500]
clothing=[5000,6000,7500,8200]
home_appliances=[3000,7500,8200,9000]

plt.figure(figsize=(10,6)) # set the figure size ,width and height

plt.stackplot(quarters,electronics,clothing,home_appliances,labels=['Electronics','Clothing','Home Appliance'])

plt.xlabel('Quarters')
plt.ylabel('Sales ($)')
plt.title('Product Catagory Sales over Quarters')
plt.legend()

plt.show()
```

```python
# 3D plots
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)

z=np.sin(np.sqrt(x**2+y**2))
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
# sub plot
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)

fig,axs=plt.subplots(2,2,figsize=(10,8))

axs[0,0].plot(x,y1,'r') #Top-left subplot
axs[0,0].set_title('sine Wave') 
axs[0,1].plot(x,y2,'b') #top-right subplot
axs[0,1].set_title('Cosine Wave')
axs[1,0].plot(x,y3,'g') #botton-left subplot
axs[1,0].set_title('tangent Wave')
axs[1,1].plot(x,y4,'m') #bottom-right subplot
axs[1,1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
